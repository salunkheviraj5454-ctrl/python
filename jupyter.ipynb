```python


a = int(input("Enter the first number: "))
b = int(input("Enter the second number: "))
c = int(input("Enter the third number: "))

if a > b and a > c:
    print("a is greatest:", a)

elif b > a and b > c:
    print("b is greatest:", b)

else:
    print("c is greatest:", c)

Enter the first number: 5
Enter the second number: 10
Enter the third number: 15
c is greatest: 15


num = int(input("Enter the number: "))

if num % 2 == 0:
    print("The number is even:", num)
else:
    print("The number is odd:", num)

Enter the number: 8
The number is even: 8


ch = input("Enter a character: ")

if ch.isupper():
    print("Uppercase character")
elif ch.islower():
    print("Lowercase character")
else:
    print("Not an alphabet")

Enter a character: Viraj
Not an alphabet


value = input("Enter the input: ")

if value.isdigit():
    print("The input value is digit:", value)

elif value.isalpha():
    print("The input value is alphabet:", value)

else:
    print("The input value is special character:", value)

Enter the input: 7
The input value is digit: 7
```
